```python
import pandas as pd
```

```python
d1={"Name":["John","Alice","Bob","Emma"],
"Age":[20,22,12,21],
"Gender":["Male","Female","Male","Female"],
"Marks":[88,92,75,89],
"City":["New York","Los Angeles","Chicago","San Fransico"]}
vr=pd.DataFrame(d1)
```

```python
vr.query("Gender=='Female'")
```

```python
vr[["Name","Marks"]].query("Marks>70")
```

```python
vr.loc[vr["Name"]=="Emma","City"]="San Fransico"
vr
```

```python
vr["Makrs"]=vr["Marks"]*1.1
vr
```

```python
def status(Marks):
    if Marks>=40:
        return "Pass"
    else:
        return "Fail"
vr["Status"]=vr["Marks"].apply(status)
vr
```

```python
import numpy as np 
vr.loc[vr["Name"]=="Bob","Marks"]=np.nan
vr["Marks"]=vr["Marks"].fillna(0) # fillna used to fill null values
vr
```

```python
vr.dropna(inplace=True)
vr.reset_index(drop=True,inplace=True) #dropna used to drop null values
vr
```
